# Dataset visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_map)
```

```python
# 1. Simple Line Plot
plt.figure(figsize=(10, 6))
average_sepal_length = iris_df.groupby('species')['sepal length (cm)'].mean()
average_sepal_length.plot(kind='line', marker='o', color='teal')
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.grid(True)
plt.show()
```

```python
# 2. Simple Scatter Plot
plt.figure(figsize=(10, 6))
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species)
plt.title('Scatter Plot of Sepal Dimensions by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Histogram
plt.figure(figsize=(10, 6))
plt.hist(iris_df['petal length (cm)'], bins=20, color='purple', edgecolor='black')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 4. Bar Chart
species_count = iris_df['species'].value_counts()
plt.figure(figsize=(10, 6))
species_count.plot(kind='bar', color=['green', 'orange', 'blue'])
plt.title('Count of Each Species in the Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()

# 5. Pair Plot (Complex)
plt.figure(figsize=(10, 8))
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

# 6. Heatmap (Complex)
plt.figure(figsize=(10, 6))
correlation_matrix = iris_df.iloc[:, :-1].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Iris Features')
plt.show()

# 7. Box Plot (Complex)
plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=iris_df)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.grid(True)
plt.show()

# 8. Violin Plot (Complex)
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='petal width (cm)', data=iris_df)
plt.title('Violin Plot of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.grid(True)
plt.show()

# 9. KDE Plot (Complex)
plt.figure(figsize=(10, 6))
sns.kdeplot(data=iris_df, x='sepal length (cm)', hue='species', fill=True)
plt.title('Kernel Density Estimate (KDE) of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.grid(True)
plt.show()

# 10. 3D Scatter Plot (Complex)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    ax.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], subset['petal length (cm)'],
               color=colors[species], label=species, s=50)
ax.set_title('3D Scatter Plot of Iris Features')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
ax.legend()
plt.show()
```
